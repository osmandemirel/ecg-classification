 # Train, validation and test data split

```
# you need to update path
import glob
import os
import shutil

data_path = "/content/data"
train_path = "/content/data/train"
val_path = "/content/data/validation"
test_path = "/content/data/test"

test_size = 0.15
val_size = 0.10

for raw_folder in glob.glob(os.path.join(data_path, "*")):
    print(raw_folder)
    train_test_ix = int(len(glob.glob(raw_folder + os.path.sep + "*.mat")) * (1-test_size))
    train_val_ix = int(train_test_ix * (1-val_size))
    print(train_val_ix, train_test_ix)

    # train
    for raw_file in glob.glob(os.path.join(raw_folder, "*.mat"))[:train_val_ix]:
        target_path = os.path.join(train_path, raw_file.split("/")[-1])
        shutil.move(raw_file, target_path)
        raw_file = raw_file.replace(".mat", ".hea")
        target_path = target_path.replace(".mat", ".hea")
        shutil.move(raw_file, target_path)

    # validation
    for raw_file in glob.glob(os.path.join(raw_folder, "*.mat"))[train_val_ix:train_test_ix]:
        target_path = os.path.join(val_path, raw_file.split("/")[-1])
        shutil.move(raw_file, target_path)
        raw_file = raw_file.replace(".mat", ".hea")
        target_path = target_path.replace(".mat", ".hea")
        shutil.move(raw_file, target_path)

    # test
    for raw_file in glob.glob(os.path.join(raw_folder, "*.mat"))[train_test_ix:]:
        target_path = os.path.join(test_path, raw_file.split("/")[-1])
        shutil.move(raw_file, target_path)
        raw_file = raw_file.replace(".mat", ".hea")
        target_path = target_path.replace(".mat", ".hea")
        shutil.move(raw_file, target_path)
```

# Import necessary classes and methods

```
# change directory to repo for imports
%cd "/content/ecg-classification"

import matplotlib.pyplot as plt
import pandas as pd
import torch
from torch import nn
from torch.utils.data import DataLoader

from models.losses import *
from models.model import ECGNet
from src.data.dataset import ECGDataset
from src.model import eval
from src.model.Trainer import Trainer
```

# Load datasets

```
class_list_path = "/content/ecg-classification/dx_mapping_scored.csv"
df_class_list = pd.read_csv(class_list_path)
class_list = sorted(df_class_list["SNOMED CT Code"])
class_name_list = df_class_list.sort_values("SNOMED CT Code")["Dx"].values
```

```
train_path = "/content/data/train"
train_data = ECGDataset(train_path, class_list)
val_path = "/content/data/validation"
val_data = ECGDataset(val_path, class_list)

batch_size = 16
train_loader = DataLoader(train_data, batch_size=batch_size, num_workers=4)
val_loader = DataLoader(val_data, batch_size=batch_size, num_workers=4)
```

```
label_list = []
for i in range(train_data.__len__()):
    label_list.append(train_data.__getitem__(i)[1])
train_label_dist = sum(label_list)

label_list = []
for i in range(val_data.__len__()):
    label_list.append(val_data.__getitem__(i)[1])
val_label_dist = sum(label_list)
```

```
def plot_label_dist(class_list, label_dist, title="title"):
    label_dict = {str(c): int(label_dist[ix].item()) for ix, c in enumerate(class_list)}
    plt.figure(figsize=(16,8))
    plt.xticks(rotation=90)
    plt.bar(label_dict.keys(), label_dict.values(), 1, color='g')
    plt.title(title)
    plt.savefig(f"/content/{title}.png")
```

```
plot_label_dist(class_name_list, train_label_dist, "Train Label Distribution")
plot_label_dist(class_name_list, val_label_dist, "Validation Label Distribution")
```

# Define and train the model with parameters 

```
in_channels = 12
num_classes = 27
ecg_net = ECGNet(in_channels, num_classes)
```

```
dx_weights_path = "/content/ecg-classification/dx_weights.csv"
criterion = SoftDiceLossWithPenalty(dx_weights_path)
criterion = nn.MultiLabelSoftMarginLoss()

conf = {
    "number_of_epochs": 25,
    "val_every": 5,
    "optimizer": "SGD",
    "lr": 0.05,
    "batch_size": batch_size,
    "result_path": "/content",
    "device": "gpu"
}
t = Trainer(ecg_net, criterion, train_loader, val_loader, config=conf)
```

```
t.train_supervised()
```

# Evaluate test data

```
test_path = "/content/data/test"
test_data = ECGDataset(test_path, class_list)
batch_size = 16
test_loader = DataLoader(test_data, batch_size=batch_size, num_workers=4)
```

```
in_channels = 12
num_classes = 27
m = ECGNet(in_channels, num_classes)

# you need to update path
model_path = "/content/model.pth"
weights = torch.load(model_path)

m.load_state_dict(weights)
```

```
dx_weights_path = "/content/ecg-classification/dx_weights.csv"
criterion = SoftDiceLossWithPenalty(dx_weights_path)
criterion = nn.MultiLabelSoftMarginLoss()

pred, target = eval.eval(m, test_loader, criterion, batch_size)
```

```
from sklearn.metrics import multilabel_confusion_matrix, classification_report
multilabel_confusion_matrix(target, pred)
print(classification_report(target, pred))
```
